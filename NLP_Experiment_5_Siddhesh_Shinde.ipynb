```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
```

```python
import re
porter_stemmer_nltk = PorterStemmer()
lancaster_stemmer_nltk = LancasterStemmer()
snowball_stemmer_nltk = SnowballStemmer('english')
def stem_words(words,stemmer):
  cleaned_words = [re.sub(r'[^\w\s]','',word) for word in words]
  return [stemmer.stem(word) for word in cleaned_words]

def main():
  text = input("Enter a list of words separated by spaces ").strip()
  words = text.split()

  porter_stemmed_words_nltk = stem_words(words,porter_stemmer_nltk)
  lancaster_stemmed_words_nltk = stem_words(words,lancaster_stemmer_nltk)
  snowball_stemmed_words_nltk = stem_words(words,snowball_stemmer_nltk)

  print('\n Porter Stemming:')
  print('Original Words:',words)
  print('Stemmed Words:',porter_stemmed_words_nltk)

  print('\n Lancaster Stemming:')
  print('Original Words:',words)
  print('Stemmed Words:',lancaster_stemmed_words_nltk)

  print('\n Snowball Stemming:')
  print('Original Words:',words)
  print('Stemmed Words:',snowball_stemmed_words_nltk)

if __name__ == '__main__':
  main()
```

### Spacy and Stanza (modern libraries) donot support the process of Stemming as the process does not land us with a meaningful base word for a given input word so they rather rely on Lemmatization to come up with a base word that's more meaningful

## Stemming or Lemmatization using Spacy and Stanza

```python
!pip install spacy
!pip install stanza
```

```python
# spaCy
import spacy
nlp_spacy = spacy.load("en_core_web_sm")

# Stanza
import stanza
stanza.download("en")  # Only needed once
nlp_stanza = stanza.Pipeline(lang="en", processors="tokenize,mwt,pos,lemma")
```

```python
def lemmatize_spacy(words):
    doc = nlp_spacy(" ".join(words))
    return [token.lemma_ for token in doc]

def lemmatize_stanza(words):
    doc = nlp_stanza(" ".join(words))
    lemmatized = []
    for sentence in doc.sentences:
        for word in sentence.words:
            lemmatized.append(word.lemma)
    return lemmatized

def main():
    text = input("Enter a list of words separated by spaces: ").strip()
    words = text.split()

    spacy_lemmatized_words = lemmatize_spacy(words)

    stanza_lemmatized_words = lemmatize_stanza(words)

    print('\n spaCy Lemmatization:')
    print('Original Words:', words)
    print('Lemmatized Words:', spacy_lemmatized_words)

    print('\n Stanza Lemmatization:')
    print('Original Words:', words)
    print('Lemmatized Words:', stanza_lemmatized_words)

if __name__ == '__main__':
    main()
```
